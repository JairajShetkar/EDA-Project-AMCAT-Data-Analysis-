#### - ECDF (Empirical Cumulative Distribution Function) Plot:


```python
def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(data) + 1) / len(data)
    return x, y

x, y = ecdf(df['Salary'])
plt.plot(x, y, marker='.', linestyle='none')
plt.title('ECDF of Salary')
plt.xlabel('Salary (Rs)')
plt.ylabel('ECDF')
plt.margins(0.02)  # Add 2% padding to the plot
plt.show()
```

#### - Summary Statistics:

```python
mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
min_salary = df['Salary'].min()
max_salary = df['Salary'].max()
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"IQR: {IQR}")
```

#### - Outlier Detection:


```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]
print("Potential Outliers:")
outliers
```

#### - Hypothesis Testing:


#### T-Test:
#### Performing a t-test, to determine if there is a statistically significant difference in salary between different college tiers.

```python
from scipy.stats import ttest_ind
```

```python
# Perform t-test between salary for different college tiers
college_tier1_salary = df[df['CollegeTier'] == 1]['Salary']
college_tier2_salary = df[df['CollegeTier'] == 2]['Salary']
```

```python
t_stat, p_value = ttest_ind(college_tier1_salary, college_tier2_salary)
print(f"T-Test Results - T-statistic: {t_stat}, p-value: {p_value}")
```

### Observation 26: From the above visualizations and statistical tests performed we can infer that there is a statistically significant difference in salary between college tier 1 and college tier 2.

###    - Therefore, we can infer that there is strong evidence to suggest that college tier has a significant impact on salary, with one tier generally associated with higher salaries compared to the other tier.

### Q2: Relationship Between Gender and Specialization

```python
# Chi-square test for independence between Gender and Specialization
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Gender'], df['Specialization'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square test result, P-value: {p}")
```

### Observation 32: In summary, the result of the Chi-square test suggests that Gender and Specialization are not independent, and there is a statistically significant association between them.

```python
# Selecting relevant columns for correlation analysis
scores_columns = ['English', 'Logical', 'Quant', 'Salary']
scores_df = df[scores_columns]

# Calculating Pearson correlation coefficient
correlation_matrix = scores_df.corr()

# Extracting correlation between scores and salary
salary_correlation = correlation_matrix.loc['Salary', ['English', 'Logical', 'Quant']]

print("Correlation between Standardized Test Scores and Employment Salary:")
print(salary_correlation)
```
